```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.linear_model import LinearRegression
```

<!-- Radinka Rafi'ie Achmad Pradipta -->

```python
df = pd.read_csv("cars.csv")
```

```python
reg = LinearRegression()
reg.fit(df[['x1','x2']].values,df.y)
```

```python
mpl.rcParams['legend.fontsize'] = 12
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(df.x1, df.x2, df.y)
ax.legend()
ax.view_init(45,0)
plt.show()
```

```python
# membaca nilai intercept
a = reg.intercept_

# membaca nilai-nilai slope
b = reg.coef_

print("Intercept = ", a)
print('koefisien regresi = ', b[0], ' dan ', b[1])
```

```python
# membuat nilai input baru yaitu x1 dan x2
X_baru = np.array([2000, 1328]).reshape(1,2)

# memprediksi y (berat ikan) berdasarkan input x1 dan x2 di atas
prediksi_Y = reg.predict(X_baru)

print('Hasil prediksi untuk x1=2000 dan x2=1328 yaitu ', prediksi_Y)
```
